```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
print('hi')
```

```python
sentences = [
    "Parinita's nickname is Tina",
    "Parinita loves coding in Python and JavaScript",
    "Parinita enjoys machine learning and data science",
    "Parinita likes building web applications",
    "Parinita is from Bengaluru India",
    "Parinita studied computer science",
    "Parinita loves reading books in her free time",
    "Parinita is passionate about AI research",
    "Parinita enjoys traveling and photography",
    "Parinita makes amazing ramen noodles"
]

df = pd.DataFrame(sentences, columns=["text"])
df
```

```python
vectorizer = TfidfVectorizer()

# Convert text → vectors
tfidf_matrix = vectorizer.fit_transform(df['text'])
tfidf_matrix
```

```python
def cosine_similarity_search(query, tfidf_matrix, df):
    # Convert query → vector
    query_vec = vectorizer.transform([query]) #tokenization+vocab-matching+tfidf for query words

    # Compute cosine similarity (dot product normalized)
    similarity_scores = cosine_similarity(query_vec, tfidf_matrix).flatten() 
    best_idx = similarity_scores.argmax() #index of the sentence that has the highest cosine similarity score with your query.

    return df['text'][best_idx], similarity_scores[best_idx]
    #.flatten() converts a multi-dimensional array into a 1-dimensional array.
```

```python
query = "what did Parinita passionate about"

result, score = cosine_similarity_search(query, tfidf_matrix, df)

print(" Query:", query)
print(" Most similar sentence:", result)
print(" Similarity score:", round(score, 3))
```
